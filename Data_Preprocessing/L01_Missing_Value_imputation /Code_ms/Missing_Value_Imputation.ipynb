## Missing Value Imputation

```python
# percent null values each column
null_val = df.isnull().sum()/df.shape[0]*100
null_val
```

```python
# columns have null values more than to 17%
col_drop = null_val[null_val > 17].keys()
col_drop
```

```python
drop_col = df.drop(columns=col_drop)
```

```python
drop_col.shape
```

```python
sns.heatmap(data=drop_col.isnull())
```

```python
clean_da = drop_col.dropna()
clean_da.shape
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(clean_da.isnull());
```

```python
clean_da.isnull().sum().sum()
```

## Data Visualization

```python
num_var = clean_da.select_dtypes(include=['float64', 'int64']).columns
num_var
```

```python
sns.distplot(df['MSSubClass'], color='c');
sns.distplot(clean_da['MSSubClass']);
```

```python
plt.figure(figsize=(25,25))
for i,var in enumerate(num_var):
    plt.subplot(10,4,i+1)
    sns.distplot(df[var],bins=20)
    sns.distplot(clean_da[var], bins=20)
```

## Compare original and cleaning object values 

```python
clean_da.select_dtypes(include='object').columns
```

```python
pd.concat([df['MSZoning'].value_counts()/df.shape[0]*100, 
           clean_da['MSZoning'].value_counts()/clean_da.shape[0]*100], 
           axis=1, keys=['MSZoning_Org', 'MSZoning_Cln'])
```

```python
#percent ratio Origin and Clean values
def per_col_val(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100, 
           clean_da[var].value_counts()/clean_da.shape[0]*100], 
           axis=1, keys=[var+'_Org', var+'_Cln'])
```
