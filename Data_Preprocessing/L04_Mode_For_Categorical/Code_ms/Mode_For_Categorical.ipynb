# [Data Preprocessing]
# Using The Measure Of Central Tendency (Mode) For Categorical Value/Data

## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv('./DATA INPUT/train.csv')
df.head()
```

## Data Preprocessing

```python
df.shape
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
percent_null = df.isnull().sum()/df.shape[0]*100
percent_null
```

```python
col_drop = percent_null[percent_null>20].keys()
col_drop
```

```python
df_drop = df.drop(columns=col_drop)
df_drop.head()
```

```python
df_drop.shape
```

```python
df_cat = df_drop.select_dtypes(include='object')
df_cat.columns
```

```python
miss_cat_per = df_cat.isnull().mean()*100
miss_cat_per
```

```python
miss_cat = miss_cat_per[miss_cat_per>0].keys()
miss_cat
```

```python
df_cat.shape
```

```python
df_cat['BsmtQual'].mode()
```

```python
df_cat['BsmtQual'].value_counts()
```

```python
df_cat['BsmtQual'].fillna(df_cat['BsmtQual'].mode()[0])
```

```python
# fill new values by mode()
for var in miss_cat:
    df_cat[var].fillna(df_cat[var].mode()[0], inplace=True)
    print(var,'=', df_cat[var].mode()[0])
```

```python
df_cat.isnull().sum().sum()
```

## Data Distribution

```python
plt.figure(figsize=(16,9))
for i, var in enumerate(miss_cat):
    plt.subplot(3,4,i+1)
    plt.hist(df_cat[var], label='Imput')
    plt.hist(df_drop[var].dropna(), label='Oroginal')
    plt.legend();
```
