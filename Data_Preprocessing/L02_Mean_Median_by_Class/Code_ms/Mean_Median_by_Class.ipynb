## Import Libraries

## Data Preprocessing

```python
df_copy = clean_dt.copy()
for var_class in clean_dt['MSZoning'].unique():
    df_copy.update(clean_dt[clean_dt.loc[:,'MSZoning']=='RL']['LotFrontage'].replace(np.nan,clean_dt[clean_dt['MSZoning']=='RL']['LotFrontage'].mean()))
```

```python
# Mean
df_mean_copy = clean_dt.copy()
miss_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_var = ['MSZoning', 'Exterior2nd', 'KitchenQual']
for cat_vars, miss_vars in zip(cat_var,miss_var):
    for var_class in clean_dt[cat_vars].unique():
        df_mean_copy.update(clean_dt[clean_dt.loc[:,cat_vars]==var_class][miss_vars].replace(np.nan,clean_dt[clean_dt[cat_vars]==var_class][miss_vars].mean()))
```

```python
df_mean_copy.isnull().sum()[miss_var]
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(miss_var):
    plt.subplot(2,2,i+1)
    sns.distplot(clean_dt[var], bins=20, label="Original",)
    sns.distplot(df_mean_copy[var], bins=20, label="Mean",)
    plt.legend();
```

```python
# Median
df_median_copy = clean_dt.copy()
miss_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_var = ['MSZoning', 'Exterior2nd', 'KitchenQual']
for cat_vars, miss_vars in zip(cat_var,miss_var):
    for var_class in clean_dt[cat_vars].unique():
        df_median_copy.update(clean_dt[clean_dt.loc[:,cat_vars]==var_class][miss_vars].replace(np.nan,clean_dt[clean_dt[cat_vars]==var_class][miss_vars].median()))
```

```python
df_median_copy.isnull().sum()[miss_var]
```

```python
clean_dt.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(miss_var):
    plt.subplot(2,2,i+1)
    sns.distplot(clean_dt[var], bins=20, label="Original")
    sns.distplot(df_mean_copy[var], bins=20, label="Mean")
    sns.distplot(df_median_copy[var], bins=20, label='Median')
    plt.legend()
```

```python
clean_dt.update(df_median_copy)
```
